```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df=pd.read_csv("netflix.csv")
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df['Date']=pd.to_datetime(df['Date'])
df=df.set_index('Date')
df.head()
```

# Volume of Stock Trade

```python
sns.lineplot(x=df.index,y=df['Volume'],label='Volume')
plt.title('volume of stock verses time')
```

```python
df.plot(y=['High','Open','Close'],title='Netflix')
```

# DAY, MONTH ,YEAR WISE GRAPH WITH VOLUME IN Y AXIS

```python
fig,(axis1,axis2,axis3)=plt.subplots(3,figsize=(15,10))
df.groupby(df.index.day).mean().plot(y='Volume', ax=axis1, xlabel='Day')
df.groupby(df.index.month).mean().plot(y='Volume' ,ax=axis2, xlabel='Month')
df.groupby(df.index.year).mean().plot(y='Volume' ,ax=axis3, xlabel='Year')
```
